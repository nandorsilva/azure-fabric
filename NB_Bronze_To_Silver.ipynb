```python
from pyspark.sql.functions import round, col, dayofmonth, month, year, to_date, quarter, substring, when, regexp_replace
```

```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
bronze_table_path ="abfss://88d42a16-2c0a-495a-97db-34b2c41a2439@onelake.dfs.fabric.microsoft.com/10aa08da-e1b2-4c37-9689-6d5e7643f00c/Tables/dbo/produtos"

df = spark.read.format("delta").load(bronze_table_path)
```

```python
display(df)
```

```python
df_transformed = (df
    .withColumn("dia", dayofmonth(col("date")))
    .withColumn("mes", month(col("date")))
    .withColumn("ano", year(col("date")))
    .withColumn("quarter", quarter(col("date")))
    .withColumn("hora", substring(col("time"), 1, 2).cast("int"))
    .withColumn("minuto", substring(col("time"), 4, 2).cast("int"))
      # período do dia
    .withColumn(
        "periodo",
        when((col("hora") >= 5) & (col("hora") < 12), "Morning")
        .when((col("hora") >= 12) & (col("hora") < 17), "Afternoon")
        .when((col("hora") >= 17) & (col("hora") < 21), "Evening")
        .otherwise("Night")
    )    
    )
```
